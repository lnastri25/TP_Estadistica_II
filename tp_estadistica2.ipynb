# Albúm de Figuritas

### Una persona desea llenar un album de 500 figuritas. Para ello comprará paquetes de 5 figuritas hasta llenarlo, las figuritas repetidas se descartan. Considerar que en cada paquete NO hay figuritas repetidas. Se quiere estudiar la cantidad de paquetes necesarios para llenar el album.

## 1) Simular el llenado del album 10000 veces.
## 2) Hallar la media, mediana y desvío estándar de la cantidad de paquetes necesarios para llenar el album.
## 3)  Realizar un histograma para tener una aproximacion a la distribucion de la cantidad de paquetes necesarios para llenar el album.
## 4) Repetir los ejercicios anteriores si los paquetes son de 6, 7 u 8 figuritas

___

```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
```

```python
def llenar_album(figus_total, figus_paquete):
    album = np.zeros(figus_total)
    paquetes = 0
    figus_pegadas = 0
    while figus_pegadas < figus_total:
        paquete = np.random.randint(0, figus_total, figus_paquete)
        paquetes += 1
        for i in paquete:
            if album[i] == 0:
                album[i] = 1
                figus_pegadas += 1
    return paquetes
```

```python
def comprar_paquetes(figus_total, figus_paquete, n_repeticiones):
    lista = []
    for i in range(n_repeticiones):
        lista.append(llenar_album(figus_total, figus_paquete))
    return lista
```

```python
def graficar_histo(lista, figus_total, figus_paquete):
    plt.hist(lista, bins=50, density=True)
    plt.xlabel('Cantidad de paquetes')
    plt.ylabel('Probabilidad')
    plt.title(f'Histograma para {figus_total} figuritas y paquetes de {figus_paquete}')
    plt.show()
```

```python
def main():
    figus_total = 500
    figus_paquete = 5
    n_repeticiones = 10000
    lista = comprar_paquetes(figus_total, figus_paquete, n_repeticiones)
    print(f'Media: {np.mean(lista)}')
    print(f'Mediana: {np.median(lista)}')
    print(f'Desvio: {np.std(lista)}')
    graficar_histo(lista, figus_total, figus_paquete)

    figus_paquete = 6
    lista = comprar_paquetes(figus_total, figus_paquete, n_repeticiones)
    print(f'Media: {np.mean(lista)}')
    print(f'Mediana: {np.median(lista)}')
    print(f'Desvio: {np.std(lista)}')
    graficar_histo(lista, figus_total, figus_paquete)

    figus_paquete = 7
    lista = comprar_paquetes(figus_total, figus_paquete, n_repeticiones)
    print(f'Media: {np.mean(lista)}')
    print(f'Mediana: {np.median(lista)}')
    print(f'Desvio: {np.std(lista)}')
    graficar_histo(lista, figus_total, figus_paquete)

    figus_paquete = 8
    lista = comprar_paquetes(figus_total, figus_paquete, n_repeticiones)
    print(f'Media: {np.mean(lista)}')
    print(f'Mediana: {np.median(lista)}')
    print(f'Desvio: {np.std(lista)}')
    graficar_histo(lista, figus_total, figus_paquete)


if __name__ == "__main__":

    main()
```

___

## Conclusión:
### Se puede observar que a medida que aumenta la cantidad de figuritas por paquete, la media y la mediana van aumentando, mientras que el desvío estándar va disminuyendo. Esto se debe a que a medida que aumenta la cantidad de figuritas por paquete, entonces la cantidad de paquetes necesarios para llenar el álbum disminuye, y por lo tanto la dispersión de los datos también disminuye. A medida que va suboendo la cantidad de figuritas por paquete, la distribución de la cantidad de paquetes necesarios para llenar el álbum se va acercando a una distribución normal.
