# Albúm de Figuritas

### Una persona desea llenar un album de 500 figuritas. Para ello comprará paquetes de 5 figuritas hasta llenarlo, las figuritas repetidas se descartan. Considerar que en cada paquete NO hay figuritas repetidas. Se quiere estudiar la cantidad de paquetes necesarios para llenar el album.

## 1) Simular el llenado del album 10000 veces.
## 2) Hallar la media, mediana y desvío estándar de la cantidad de paquetes necesarios para llenar el album.
## 3)  Realizar un histograma para tener una aproximacion a la distribucion de la cantidad de paquetes necesarios para llenar el album.
## 4) Repetir los ejercicios anteriores si los paquetes son de 6, 7 u 8 figuritas

___

```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0) # Seteo la semilla para que los resultados sean reproducibles.
```

```python
def llenar_album(figus_total, figus_paquete): # Defino la funcion que llena el album.
    album = np.zeros(figus_total) # Creo un album vacio
    paquetes = 0 # Inicializo el contador de paquetes.
    figus_pegadas = 0 # Inicializo el contador de figuritas pegadas.
    while figus_pegadas < figus_total: # Mientras no este lleno el album, hace lo siguiente:
        paquete = np.random.randint(0, figus_total, figus_paquete) # uso random.randint para generar un paquete de figuritas al azar.
        paquetes += 1 # Cuento la cantidad de paquetes que voy comprando.
        for i in paquete: # Recorro el paquete y voy llenando el album.
            if album[i] == 0: # Si la figurita no esta en el album, la pego.
                album[i] = 1 # Pego la figurita.
                figus_pegadas += 1 # Cuento la cantidad de figuritas pegadas.
    return paquetes
```

```python
def comprar_paquetes(figus_total, figus_paquete, n_repeticiones): # Defino la funcion que repite el experimento.
    lista = [] # Creo una lista vacia.
    for i in range(n_repeticiones): # Repito n_repeticiones veces el experimento.
        lista.append(llenar_album(figus_total, figus_paquete)) # Guardo la cantidad de paquetes que compre en cada repeticion.
    return lista
```

```python
def graficar_histo(lista, figus_total, figus_paquete):
    plt.hist(lista, bins=50, density=True)
    plt.xlabel('Cantidad de paquetes')
    plt.ylabel('Porcentaje de personas')
    plt.title(f'Histograma para {figus_total} figuritas y paquetes de {figus_paquete}')
    plt.show()
```

```python
def main():
    figus_total = 500
    figus_paquete = 5
    n_repeticiones = 10000
    lista = comprar_paquetes(figus_total, figus_paquete, n_repeticiones)
    print(f'Media: {np.mean(lista)}')
    print(f'Mediana: {np.median(lista)}')
    print(f'Desvio: {np.std(lista)}')
    graficar_histo(lista, figus_total, figus_paquete)

    figus_paquete = 6
    lista = comprar_paquetes(figus_total, figus_paquete, n_repeticiones)
    print(f'Media: {np.mean(lista)}')
    print(f'Mediana: {np.median(lista)}')
    print(f'Desvio: {np.std(lista)}')
    graficar_histo(lista, figus_total, figus_paquete)

    figus_paquete = 7
    lista = comprar_paquetes(figus_total, figus_paquete, n_repeticiones)
    print(f'Media: {np.mean(lista)}')
    print(f'Mediana: {np.median(lista)}')
    print(f'Desvio: {np.std(lista)}')
    graficar_histo(lista, figus_total, figus_paquete)

    figus_paquete = 8
    lista = comprar_paquetes(figus_total, figus_paquete, n_repeticiones)
    print(f'Media: {np.mean(lista)}')
    print(f'Mediana: {np.median(lista)}')
    print(f'Desvio: {np.std(lista)}')
    graficar_histo(lista, figus_total, figus_paquete)


if __name__ == "__main__":

    main()
```

## Conclusión:

### Al analizar el efecto de la cantidad de figuritas por paquete en el proceso de llenar un álbum, pudimos notar un patrón claro. A medida que aumenta la cantidad de figuritas por paquete, la media, la mediana y el desvío estándar disminuyen. Esto se debe a que se necesitan menos paquetes en promedio para completar el álbum, lo que reduce la dispersión en los datos. Esto último, también lleva a que la distribución de la cantidad de paquetes necesarios se parezca más a una distribución normal.

### En lo que respecta a los gráficos, podemos destacar que la mediana representa el punto donde se encuentra la probabilidad más alta de completar el álbum, y esta probabilidad decrece a medida que aumenta o disminuye la cantidad de paquetes comprados. En otras palabras, la mediana indica la cantidad de paquetes que con mayor frecuencia resulta en la finalización del álbum.

### En resumen, la elección de la cantidad de figuritas por paquete tiene un impacto significativo en la probabilidad de éxito al llenar un álbum. Aumentar esta cantidad tiende a optimizar el proceso, reduciendo la variabilidad/dispersión y acercando la distribución de resultados a una forma más predecible. La mediana cobra un rol importante para determinar la cantidad de paquetes más efectiva en la búsqueda de completar el álbum, ya que se sitúa en el punto de mayor probabilidad de éxito.
